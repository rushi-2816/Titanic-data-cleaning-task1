```python
import pandas as pd

# Try reading it as Excel
try:
    df = pd.read_excel("Titanic-Dataset.csv.xls")
except:
    # Fallback: try as CSV
    df = pd.read_csv("Titanic-Dataset.xls")

# Show the first 5 rows
df.head()
```

```python
# Step 1: Basic Info
print("ðŸ” Dataset Info:")
print(df.info())

print("\nðŸ§¹ Missing Values:")
print(df.isnull().sum())

# Step 2: Handling Missing Values
# Fill Age with median
df['Age'].fillna(df['Age'].median(), inplace=True)

# Fill Embarked with mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Drop Cabin due to too many missing values
if 'Cabin' in df.columns:
    df.drop(columns=['Cabin'], inplace=True)

# Step 3: Encoding Categorical Variables
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])  # male:1, female:0

# One-hot encoding for 'Embarked'
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)

# Step 4: Feature Scaling (Age, Fare)
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])

# Step 5: Detecting and Removing Outliers
import seaborn as sns
import matplotlib.pyplot as plt

# Visualize outliers in 'Fare'
sns.boxplot(x=df['Fare'])
plt.title("Boxplot of Fare")
plt.show()

# Remove outliers using IQR method
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1

# Keep only data within IQR range
df = df[(df['Fare'] >= Q1 - 1.5 * IQR) & (df['Fare'] <= Q3 + 1.5 * IQR)]

# Final Preview
print("\nâœ… Cleaned Dataset Preview:")
print(df.head())

# Step 6: Save the Cleaned File
df.to_csv("titanic_cleaned.csv", index=False)
print("\nðŸ’¾ Saved cleaned file as 'titanic_cleaned.csv'")
```
